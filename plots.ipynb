### Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
x1 = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205])
y1 = np.array([55, 58, 60, 62, 64, 66, 68, 70, 72, 74])

x2 = np.array([150, 155, 160, 165, 170, 175, 180, 195, 200, 205])
y2 = np.array([50, 52, 54, 56, 58, 64, 66, 68, 70, 72])

plt.scatter(x1, y1, c="blue", label="Group1")
plt.scatter(x2, y2, c="red", label="Group2")

plt.title("Comparison of Height vs Weight between two groups")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

plt.legend()
plt.show()
```

```python
x = np.array([3, 12, 9, 20, 5, 18, 22, 11, 27, 16])
y = np.array([95, 55, 63, 77, 89, 50, 41, 70, 58, 83])

a = [20, 50, 100, 200, 500, 1000, 60, 90, 150, 300] # size
b = ['red', 'green', 'blue', 'purple', 'orange', 'black', 'pink', 'brown', 'yellow', 'cyan'] # color

plt.scatter(x, y, s=a, c=b, alpha=0.6, edgecolors='w', linewidths=3)
plt.title("Scatter Plot with Varying Colors and Sizes")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
sizes = [30, 80, 150, 200, 300]  # Bubble sizes

plt.scatter(x, y, s=sizes, alpha=0.5, edgecolors='blue', linewidths=2)
plt.title("Bubble Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.random.randint(50, 150, 100)
y = np.random.randint(50, 150, 100)
colors = np.random.rand(100)
sizes = 200

plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7)
plt.colorbar(label='Color scale')
plt.title("Scatter Plot with Colormap and Colorbar")
plt.show()
```

```python
x = np.random.randint(50, 150, 100)
y = np.random.randint(50, 150, 100)
colors = np.random.rand(100)
sizes = 20 * np.random.randint(10, 100, 100)

plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7)
plt.colorbar(label='Color scale')
plt.title("Scatter Plot with Colormap and Colorbar")
plt.show()
```

```python
plt.scatter(x, y, marker='^', color='magenta', s=100, alpha=0.7)
plt.title("Scatter Plot with Triangle Markers")
plt.show()
```

## matplotlib.pyplot.colorbar() function
The matplotlib.pyplot.colorbar() function in Matplotlib is used to add a color scale (colorbar) to a plot, providing a visual key to interpret the relationship between data values and colors in colormapped plots.

```python
p = [100, 200, 150, 23, 30, 50, 156, 32, 67, 89]  # Purchase counts  
l = [50, 70, 100, 10, 10, 34, 56, 18, 35, 45]  # Likes  
r = [1, 0.53, 2, 0.76, 0.5, 2.125, 0.56, 1.28, 1.09, 1.02]  # Like/Dislike ratio

# scatterplot
plt.scatter(p, l, c=r, cmap="summer")

plt.colorbar(label="Like/Dislike Ratio", orientation="horizontal") # add Colorbar
plt.xlabel("Purchases")
plt.ylabel("Likes")
plt.title("Purchases vs Likes") 
plt.show()
```

```python
# 2x2 grid  
fig, axes = plt.subplots(2, 2, figsize=(8, 8))  

# random images
for ax in axes.flat:  
    im = ax.imshow(np.random.random((10, 10)), vmin=0, vmax=1)  

# shared colorbar  
plt.colorbar(im, ax=axes.ravel().tolist())  
plt.show()
```

```python
import matplotlib as mpl

x = np.linspace(0, 5, 100)
cmap = plt.get_cmap('jet', 7)

fig, ax = plt.subplots(figsize=(8, 6))
for i, n in enumerate(np.linspace(0, 2, 7)):
    ax.plot(x, x * i + n, color=cmap(i))

sm = mpl.cm.ScalarMappable(cmap=cmap, norm=mpl.colors.Normalize(0, 2))
plt.colorbar(sm, ax=ax, ticks=np.linspace(0, 2, 7), label="Color Mapping")

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Colorbar for Non-Mappable Object")
plt.show()
```

```python
data = np.random.rand(10, 10)

fig, ax = plt.subplots()
cax = ax.imshow(data, cmap='plasma')

# Adding a customized colorbar
cbar = plt.colorbar(cax)
cbar.set_label('Intensity', rotation=270, labelpad=15)
cbar.ax.tick_params(labelsize=10)

plt.title("Customized Colorbar Example")
plt.show()
```

```python
data = np.random.rand(10, 10)

fig, ax = plt.subplots()
cax = ax.imshow(data, cmap='coolwarm')

# Adding a Horizontal Colorbar
plt.colorbar(cax, orientation='horizontal', label='Temperature Scale')
plt.title("Horizontal Colorbar Example")
plt.show()
```

## matplotlib.pyplot.imshow()
The matplotlib.pyplot.imshow() function is used to display data as an image on a 2D regular raster. It is a core function within the matplotlib.pyplot module for visualizing images, matrices, or heatmaps where each value in an array corresponds to a color.

```python
a = np.random.random((10, 10))

# Display the array as an image
plt.imshow(a, cmap='viridis', interpolation='nearest')
plt.colorbar()  
plt.show()
```

#### 2D Data with imshow() and Custom Color Range

```python
from matplotlib.colors import LogNorm 
	
dx, dy = 0.015, 0.05 # Set Grid Spacing

y, x = np.mgrid[slice(-4, 4 + dy, dy), 
				slice(-4, 4 + dx, dx)] # slice(start, stop, step) defines the start, stop, and step for grid generation

z = (1 - x / 3. + x ** 5 + y ** 5) * np.exp(-x ** 2 - y ** 2) 
z = z[:-1, :-1] 
z_min, z_max = -np.abs(z).max(), np.abs(z).max() 

c = plt.imshow(z, cmap ='Greens', vmin = z_min, vmax = z_max, 
				extent =[x.min(), x.max(), y.min(), y.max()], 
					interpolation ='nearest', origin ='lower') 
plt.colorbar(c) 

plt.title('matplotlib.pyplot.imshow() function', fontweight ="bold") 
plt.show() 
```

### Line chart

```python
x = np.array([1, 2, 3, 4])  # X-axis 
y = x*2  # Y-axis [2, 4, 6, 8]

plt.plot(x, y)  
plt.xlabel("X-axis") # Label for the X-axis
plt.ylabel("Y-axis") # Label for the Y-axis
plt.title("Title") # Chart title
plt.show() 
```

#### Line Chart with Annotations

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')

# Add annotations
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)
plt.show()
```

```python
x = np.array([1, 2, 3, 4])  # X-axis 
y = x*2  # Y-axis [2, 4, 6, 8]

plt.plot(x, y, linestyle='-')  
plt.show() 

plt.plot(x, y, linestyle='-.')  
plt.show() 
```

```python
x = np.array([1, 2, 3, 4])
y = x*2

# first plot with X and Y data
plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Multiple Plots')
plt.show()
```

### Area Between Two Lines (pyplot.fill_between() function)

```python
x = np.array([1, 2, 3, 4])
y = x*2 # [2, 4, 6, 8]

plt.plot(x, y, '-')

y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```

#### Multiple bar plots

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 

IT = [12, 30, 1, 8, 22] 
ECE = [28, 6, 16, 5, 10] 
CSE = [29, 3, 24, 25, 17] 

br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

plt.bar(br1, IT, color ='r', width = barWidth, edgecolor ='grey', label ='IT') 
plt.bar(br2, ECE, color = 'b', width = barWidth, edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color = 'g', width = barWidth, edgecolor ='grey', label ='CSE')

plt.xlabel('Branch', fontweight ='bold', fontsize = 15) 
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15) 
plt.xticks([r + barWidth for r in range(len(IT))], 
           ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```

#### Stacked bar plot

```python
N = 5

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)   
width = 0.35  

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width, bottom = boys, yerr = girlStd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))

plt.show()
```

### Histogram 

```python
# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a basic histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

```python
# Histogram with Density Plot 

import seaborn as sns

data = np.random.randn(1000)

# Plots a histogram with 30 bins and a smooth density curve (KDE) using Seaborn.
sns.histplot(data, bins=30, kde = True, color='skyblue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

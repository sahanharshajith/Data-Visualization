```python
numeric_cols = ['BudgetUSD', 'US_BoxOfficeUSD', 'Global_BoxOfficeUSD', 'Opening_Day_SalesUSD', 
                'One_Week_SalesUSD', 'IMDbRating', 'RottenTomatoesScore', 'NumVotesIMDb', 
                'NumVotesRT']
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for cols in numeric_cols:
    sns.histplot(data[cols], bins=50, kde=True)
    plt.title(f"Distribution of {cols}")
    plt.show()
```

```python
data.columns
```

```python
categorial_cols = ['Genre', 'Country', 'Director', 'LeadActor']
```

```python
for cols in categorial_cols:
    data[cols].value_counts().head(20).plot(kind='bar', figsize=(10, 5))
    plt.title(f"Top 20 {cols} counts")
    plt.show()
```

```python
data.columns
```

```python
sns.scatterplot(data = data, x = 'BudgetUSD', y = 'Global_BoxOfficeUSD', alpha=0.5)
plt.title("Budget vs Global Box Office")
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
sns.scatterplot(
    data=data,
    x='IMDbRating',
    y='RottenTomatoesScore',
    s=10,
    alpha=0.3,
    linewidth=0
)
plt.title("IMDb Rating vs Rotten Tomatoes Score")
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
data.columns
```

```python
genre_year_stats = data.groupby(['Genre', 'ReleaseYear']).agg({
    "Global_BoxOfficeUSD" : "mean",
    "BudgetUSD" : "mean",
    "IMDbRating" : "mean",
    "RottenTomatoesScore" : "mean"
}).reset_index()
```

```python
genre_year_stats
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data = genre_year_stats, x = 'ReleaseYear', y = 'Global_BoxOfficeUSD', hue='Genre')
plt.title("Average Global Box Office by Genre Over Years")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data = genre_year_stats.query("Genre == 'Action'"), x = 'ReleaseYear', y = 'Global_BoxOfficeUSD', hue='Genre')
plt.title("Average Global Box Office for Action and Thriller Genres Over Years")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data = genre_year_stats.query("Genre == 'Action' or Genre == 'Thriller'"), x = 'ReleaseYear', y = 'Global_BoxOfficeUSD', hue='Genre')
plt.title("Average Global Box Office for Action and Thriller Genres Over Years")
plt.show()
```

```python
top_actors = data.groupby('LeadActor')["IMDbRating"].mean().sort_values(ascending=False).head(20)
```

### Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = ['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu']
y = [5, 6.7, 4, 6, 2, 4.9, 1.8]

ax = sns.stripplot(x=x, y=y)
ax.set(xlabel="Days", ylabel="Amount")
plt.title("Daily Spending (Custom Data)")
plt.show()
```

### Swarm Plot

```python
sns.set(style="whitegrid")
data = sns.load_dataset("iris")
sns.swarmplot(x="species", y="sepal_length", data=data)
plt.title("Swarm Plot of Sepal Length by Species")
plt.show()
```

### Barplot 

```python
df = sns.load_dataset("titanic")
sns.barplot(x="class", y="fare", data=df)
plt.show()
```

```python
# Vertical barplot 
sns.barplot(x="class", y="fare", hue="sex", data=df)
plt.show()
```

```python
# Horizontal barplot
sns.barplot(x="fare", y="class", hue="sex", data=df)
plt.show()
```

```python
# barplot in given order
sns.barplot(x = 'class', y = 'fare', data = df, order = ["Third", "Second", "First"])

plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, color='salmon')

plt.show()
```

```python
# without error bars
sns.barplot(x='class', y='fare', data=df, ci=None)

plt.show()
```

```python
sns.barplot(x='class', y='fare', hue='sex', data=df, palette='pastel', ci=None)
plt.show()
```

### Count Plot

```python
df = sns.load_dataset('tips')
sns.countplot(x='sex', data=df)
plt.show()
```

```python
sns.countplot(x='sex', data=df ,hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex', data=df)
plt.show()
```

```python
# Use different color palette attributes 
sns.countplot(x='sex', palette = "Set2", data=df)
plt.show()
```

```python
#using a color parameter
data= sns.load_dataset('titanic')

sns.countplot(x='class', data=data, hue='sex', color='salmon')
plt.show()
```

### Violinplot 

```python
sns.set(style = 'whitegrid')
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.violinplot(x='timepoint', y='signal', data=fmri)
```

```python
sns.set(style = 'whitegrid') 
sns.violinplot(x='timepoint', y='signal', hue='region', data=fmri)
```
